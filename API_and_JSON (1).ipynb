#Apa itu API?
API (Application Programming Interface) adalah antarmuka yang memungkinkan
satu program berkomunikasi dan bertukar data dengan sistem atau program lain
secara terstruktur dan aman.

Cara paling mudah memahami API adalah membayangkannya sebagai seorang pelayan restoran yang menjembatani komunikasi antara
pelanggan dan dapur.

Kamu (Pelanggan)
Membuat permintaan — misalnya meminta
data nama atau cuaca hari ini.

API(Pelayan)
Menerima permintaanmu, meneruskannya
ke server, lalu membawa kembali hasilnya
kepadamu.

Server (Dapur)
Memproses permintaan, menyiapkan data
yang diminta, dan mengembalikan respons
melalui API.
Alurnya sederhana: Kamu pesan → API teruskan → Server proses → API bawa jawaban → Kamu terima data!

#Apa itu JSON?

JSON (JavaScript Object Notation) adalah format teks standar yang digunakan untuk menyimpan dan mengirimkan data terstruktur. JSON sangat
populer karena mudah dibaca manusia dan mudah diproses oleh mesin.

```python
import json

json_string = '''{"name": "Budi",
"age": 25,
"city": "Jakarta"}'''

data = json.loads(json_string)
print(data["name"])
# Output: Budi
print(data["age"])
# Output: 25
print(data["city"])
# Output: Jakarta
```

```python
import requests
url = "https://api.agify.io?name=Budi"
response = requests.get(url)
print(response)
print(response.text)
```

```python
data = response.json()
print(data)

print(data["name"])

print(data["age"])
```

```python
import requests
nama = input("Masukkan nama: ")
url = f"https://api.agify.io?name={nama}"
response = requests.get(url)
data = response.json()
print(f"Halo, {data['name']}!")
print(f"Perkiraan umur: {data['age']} tahun")
```

```python
import requests
url = "https://api.quotable.io/random"
data = requests.get(url, verify=False).json()
print(data["content"])
print("-", data["author"])
```

```python
import requests
for i in range(3):
  data = requests.get(
      "https://api.quotable.io/random", verify=False
      ).json()
  print(f"--- Quote ke-{i+1} ---")
  print(data["content"])
  print(f"- {data['author']}")
  print()
```

```python
import requests
url = "https://quotes.liupurnomo.com/api/quotes/random"
data = requests.get(url).json()
print(data["data"]["text"])
print("-", data["data"]["author"])
```

```python
import requests
for i in range(9):
    url = "https://quotes.liupurnomo.com/api/quotes?category=motivasi&page=1&limit=10"
    data = requests.get(url).json()
    print(f"--- Quote ke-{i+1} ---")
    print(data["data"][i+1]["text"])
    print(f"- {data['data'][i+1]['author']}")
    print()
```
