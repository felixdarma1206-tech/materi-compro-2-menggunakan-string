```python
# TUGAS 1 - COMPUTER PROGRAMMING 2
# String Manipulation & Data Cleaning

# 1.DATA INPUT
data_kotor = "   #001-wahyu aji-informatika-python,java,php-aktif   "
print(f"Data Mentah: '{data_kotor}'\n")

# 2.Proses pembersihan

# 2a.Menghapus spasi untuk ujung karakter diawal
data_bersih=data_kotor.strip()
data_bersih=data_bersih.replace("#","")

# 2b.memecah string menjadi list berdasarkan tanda pemisah '-'
data_list=data_bersih.split("-")

# 2c.mengelola daftar skill
skill_list=data_list[3].split(",")
skill_dikuasai=" & ".join(skill_list)

# 2d.menghitung panjang ID dan jumlah skill
id_pendaftar=data_list[0]
panjang_id=len(id_pendaftar)
total_skill=len(skill_list)

# 3.output akhir
print('=' * 40)
print("     HASIL VERIFIKASI DATA")
print('=' * 40)

# menggunakan ljust() agar titik dua sejajar
print(f"{'ID Pendaftar':15} : {id_pendaftar}({panjang_id}digit)")
print(f"{'Nama Lengkap':15} : {data_list[1].title()}")
print(f"{'Progam Studi':15} : {data_list[2].upper()}")
print(f"{'Skill Dikuasai':15} : {skill_dikuasai}")
print(f"{'Total Skill':15} : {total_skill}")
print(f"{'Status Akun':15} : {data_list[4].capitalize()}")

print('=' * 40)
```
