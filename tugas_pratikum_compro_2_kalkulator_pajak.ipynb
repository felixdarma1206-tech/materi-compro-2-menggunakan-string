```python
# Mengimpor modul dari paket transaksi
from transaksi.diskon import hitung_diskon
from transaksi.pajak import hitung_ppn

# 1. Inisialisasi Data
harga_awal = 1000000
persen_diskon = 10

# 2. Proses Perhitungan
besaran_diskon = hitung_diskon(harga_awal, persen_diskon)
harga_setelah_diskon = harga_awal - besaran_diskon
besaran_ppn = hitung_ppn(harga_setelah_diskon)
total_bayar = harga_setelah_diskon + besaran_ppn

# Fungsi bantuan untuk format Rupiah (opsional, agar output rapi)
def format_rp(nilai):
    return f"Rp {int(nilai):,}.-".replace(',', '.')

# 3. Menampilkan ke Layar (Terminal/Output Cell)
print("=== STRUK TRANSAKSI BISNIS DIGITAL ===")
print(f"Harga Awal               : {format_rp(harga_awal)}")
print(f"Besaran Potongan Diskon  : {format_rp(besaran_diskon)}")
print(f"Harga Setelah Diskon     : {format_rp(harga_setelah_diskon)}")
print(f"Besaran PPN (11%)        : {format_rp(besaran_ppn)}")
print("-" * 40)
print(f"Total yang harus dibayar : {format_rp(total_bayar)}")
```
